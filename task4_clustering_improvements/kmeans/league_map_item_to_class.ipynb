```python
from collections import defaultdict

file_path = output_file_path

class_item_count = {}

with open(file_path, 'r') as file:
    for line in file:
        values = line.strip().split(',')
        class_name = values[1]
        item_ids = values[2:]
        for item_id in item_ids:
            if item_id not in class_item_count:
                class_item_count[item_id] = {}
            class_item_count[item_id][class_name] = class_item_count[item_id].get(class_name, 0) + 1

itemclass = {}
for item_id, class_count in class_item_count.items():
    print(f"Item: {item_id}, Classes: {class_count}")
    res = 0
    for x in class_count:
        if class_count[x] > res:
            itemclass[item_id] = x
            res = class_count[x]
```
